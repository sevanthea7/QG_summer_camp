```python
headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:128.0) Gecko/20100101 Firefox/128.0' }
response2 = requests.get( 'https://movie.douban.com/top250' , headers = headers )


response2.status_code
```

```python
content1 = response1.text
content1
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup( content1, 'html.parser' )
soup.p
```

```python
all_prices = soup.findAll( "p", attrs = { "class":"price_color" } )
for price in all_prices:
	print( price.string[ 2: ] )
```

```python
all_title = soup.findAll( "h3" )
for title in all_title:
	all_links = title.findAll( "a" )
	for link in all_links:
		print( link.string )
```

```python
for title in all_title:
	all_links = title.find( "a" )
	print( all_links.string )
```

```python
for start_num in range( 0, 250, 25 ):
	response2 = requests.get( f'https://movie.douban.com/top250?start={start_num}' , headers = headers )
	html = response2.text
	soup2 = BeautifulSoup( html, 'html.parser' )
	all_titles = soup2.findAll( "span", attrs = { 'class' : 'title' } )
	for title in all_titles:
		title = title.string
		if '/' not in title:
			print( title )
```
