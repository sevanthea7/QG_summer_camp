```python
from PIL import Image
import io

# 打开图片
image = Image.open( "../data/test2.png" )

if image.mode == 'RGBA':
    image = image.convert( 'RGB' )


byte_array = io.BytesIO()


image.save(byte_array, format='JPEG')
byte_data = byte_array.getvalue()
byte_array.close()


print( byte_data ) 
print( len( byte_data ) ) 
```
